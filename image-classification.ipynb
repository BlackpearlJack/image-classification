```python
import numpy as np
import cv2
import plotly.express as px
import plotly.graph_objects as go
```

```python
img = cv2.imread('./assets/Demi_Lovato/download (1).jpg')
img.shape
```

```python
# Display the image
fig = px.imshow(img)
fig.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
gray
```

```python
px.imshow(gray, color_continuous_scale="gray")
```

```python
face_cascade = cv2.CascadeClassifier('./haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./haarcascades/haarcascade_eye.xml')

faces = face_cascade.detectMultiScale(gray, 1.3, 5)
faces
```

```python
(x,y,w,h) = faces[0]
x,y,w,h
```

```python
face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
px.imshow(face_img)
```

```python
cv2.destroyAllWindows()
for (x,y,w,h) in faces:
    face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew, ey+eh),(0,255,0),2)
        
fig = px.imshow(face_img, color_continuous_scale="gray")
fig.show()
```

```python
px.imshow(roi_color, color_continuous_scale="gray")
```

```python
def get_cropped_image_if_2_eyes(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            return roi_color
```

```python
original_image = cv2.imread('./assets/Demi_Lovato/download (1).jpg')
px.imshow(original_image)
```

```python
cropped_image = get_cropped_image_if_2_eyes('./assets/Demi_Lovato/download (1).jpg')
px.imshow(cropped_image)
```

```python
cropped_image_no_2_eyes = get_cropped_image_if_2_eyes('assets/Isabel_Allende/download (1).jpg')
cropped_image_no_2_eyes
```
